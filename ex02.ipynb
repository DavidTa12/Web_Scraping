```python
import numpy
import pandas as pd
import pickle as pickle
import sched, time
from datetime import datetime, timedelta
from IPython.display import clear_output
import os

data_file = 'data.pkl'
s = sched.scheduler(time.time, time.sleep)


def serialize_leboncoin_dataframe(data_df):
    data_df.to_pickle(data_file)

def feed_pickle_file_with_new_data():
    clear_output(wait=True)
    #retrieve data from existing pickle file if already exists
    if os.path.isfile(data_file):
        df_loaded = pickle.load(open(data_file, 'rb'))
        print('INIT PICKLE FILE CONTENT:', datetime.now() )
        display(df_loaded)

    #if it does not exist, initialize empty DF
    else:
        df_loaded = pd.DataFrame()

    # retrieve newest data from leboncoin
    new_data_df = extract_leboncoin_data()

    #merge data from pickle file with newest data + delete duplicates + sort by posting_date
    merged_dfs = df_loaded.append(new_data_df).drop_duplicates().reset_index(drop=True).sort_values(by=['posting_date'], ascending=False)


    # write updated dataframe into pickle file
    serialize_leboncoin_dataframe(merged_dfs)

    print('FINAL PICKLE FILE CONTENT:', datetime.now())
    display(merged_dfs)

    # to schedule the function execution: every minute
    s.enter(60, 1, feed_pickle_file_with_new_data)

def main():
    feed_pickle_file_with_new_data()

    # to schedule the function execution: every minute
    s.enter(60, 1, feed_pickle_file_with_new_data)
    s.run()
    

main()
```
